## I-1 Additions

```python
tmp_add = df_w_transloc.loc[:,'add_10p':'add_xq']
tmp_add = pd.concat([tmp_add,df_w_transloc.loc[:,'plus1':'plusy']],1)
to_keep = []
for i in range(1,23):
    tmp_add['add_'+str(i)] = tmp_add['add_'+str(i)+'p']+tmp_add['add_'+str(i)+'q']+tmp_add['plus'+str(i)]
    freq = 100*sum(tmp_add['add_'+str(i)])/tmp_add.shape[0]
    print ('Chromosome : '+str(i)+' with Freqency(%)', freq)
    to_keep +=['add_'+str(i) if freq >=2 else '']
    
tmp_add['add_x'] = tmp_add['add_xp']+tmp_add['add_xq']+tmp_add['plusx']
freq = 100*sum(tmp_add['add_x'])/tmp_add.shape[0]
print ('Chromosome : x with Freqency(%)', freq)
to_keep +=['add_x' if freq >=2 else '']

freq = 100*sum(tmp_add['plusy'])/tmp_add.shape[0]
print ('Chromosome : y with Freqency(%)', freq)
to_keep +=['plusy' if freq >=2 else '']


to_keep= [t for t in to_keep if t!='']
print()
print('We keep : '+ str(to_keep))
tmp_add = tmp_add[to_keep]
tmp_add[tmp_add>1]=1
```

## I-2 Deletions

```python
tmp_del = df_w_transloc.loc[:,'del_10p':'del_xq']
tmp_del = pd.concat([tmp_del,df_w_transloc.loc[:,'minus1':'minusy']],1)
tmp_del.drop(['minus7.'],1,inplace=True)
to_keep = []
list_p_missing = ['4', '14', '20', '21', '22']
list_with_p = ['1', '2', '3', '5', '6', '7', '8', '9', '10', '11', '12', '13', '15', '16', '17', '18', '19',  ]
for l in list_p_missing:
    tmp_del['del_'+l] = tmp_del['del_'+l+'q'] + tmp_del['minus'+l]
    freq = 100*sum(tmp_del['del_'+l])/tmp_del.shape[0]
    print ('Chromosome : '+l+' with Freqency(%)', freq)
    to_keep +=['del_'+l if freq >=2 else '']
    
for l in list_with_p:
    tmp_del['del_'+l] = tmp_del['del_'+l+'p'] + tmp_del['del_'+l+'q'] + tmp_del['minus'+l]
    freq = 100*sum(tmp_del['del_'+l])/tmp_del.shape[0]
    print ('Chromosome : '+l+' with Freqency(%)', freq)
    to_keep +=['del_'+l if freq >=2 else '']
    
tmp_del['del_x'] = tmp_del['del_xq'] + tmp_del['minusx'] 
freq = 100*sum(tmp_del['del_x'])/tmp_del.shape[0]
to_keep +=['del_x' if freq >=2 else '']
print ('Chromosome : x with Freqency(%)', 100*sum(tmp_del['del_x'])/tmp_del.shape[0])

freq = 100*sum(tmp_del['minusy'])/tmp_del.shape[0]
to_keep +=['minusy' if freq >=2 else '']
print ('Chromosome : y with Freqency(%)', 100*sum(tmp_del['minusy'])/tmp_del.shape[0])

to_keep= [t for t in to_keep if t!='']
print()
print('We keep : '+ str(to_keep))
tmp_del = tmp_del[to_keep]
tmp_del[tmp_del>1]=1
```

### A) Translocation features Additions

#### We follow the rules that we have set up

```python
t = df_w_transloc.loc[:,'add_10p':'add_xq']
t = pd.concat([t,df_w_transloc.loc[:,'plus1':'plusy']],1)
t.shape[1]
```

There are 70 features for the Additions: 23*2+24=70 


23: add_ chromosome 1 to 22 and chromosome x


2: p and q


24: plus chromosome 1 to 22 and x and y

#### Let's prepare the rules

```python
add = df_w_transloc.loc[:,'add_10p':'add_xq']
add = pd.concat([add,df_w_transloc.loc[:,'plus1':'plusy']],1)

lis = [1,2,3,4,5,6,7,9,10,12,13,14,15,16,17,18,19,20]  # basic rule

li = [8,11,21,22]  #rules for the ones with more than 2%

# we create the column merge to sum all of them and we drop the other columns : merge= p+q+plus
for l in lis:
    add['add_'+str(l)+'_merge']=add['add_'+str(l)+'p']+add['add_'+str(l)+'q']+add['plus'+str(l)]
    add.drop(['add_'+str(l)+'p','add_'+str(l)+'q','plus'+str(l)], axis=1,inplace=True)
    
# we create the column merge to sum the one with less than 2% frequency and we drop the columns used in the sum : merge = p+q
for l in li:
    add['add_'+str(l)+'_merge']=add['add_'+str(l)+'p']+add['add_'+str(l)+'q']
    add.drop(['add_'+str(l)+'p','add_'+str(l)+'q'], axis=1,inplace=True)
    
add['add_merge_x_y']=add['add_xp']+add['add_xq']+add['plusx']+add['plusy']    
add.drop(['add_xp','add_xq','plusx','plusy'], axis=1,inplace=True)
```

### B) Translocation features Deletions

```python
d = df_w_transloc.loc[:,'del_10p':'del_xq']
d = pd.concat([d, df_w_transloc.loc[:,'minus1':'minusy']],1)
d.drop(columns=['minus7.'],inplace=True)
d.shape[1]
```

##### There are 64 features instead of 70 because we are missing a few p in the deletions . I also deleted a weird column 'minus7.'

#### Let's prepare the rules

```python
d.columns
```

##### We can notice that the 'p' is missing for 6 of the chromosomes so we will have to handle this : 4,14,20,21,22,x

```python
deletion = df_w_transloc.loc[:,'del_10p':'del_xq']
deletion=pd.concat([deletion,df_w_transloc.loc[:,'minus1':'minusy']],1)
deletion.drop(['minus7.'],1,inplace=True)

lis = [1,2,3,6,8,10,11,12,13,15,16,19] # basic rule
lis_p_missing = [4,14,20,21,22] # when p is missing

for l in lis:
    deletion['del_'+str(l)+'_merge']=deletion['del_'+str(l)+'p']+deletion['del_'+str(l)+'q']+deletion['minus'+str(l)]
    deletion.drop(['del_'+str(l)+'p','del_'+str(l)+'q','minus'+str(l)], axis=1,inplace=True)

for l in lis_p_missing:
    deletion['del_'+str(l)+'_merge']=deletion['del_'+str(l)+'q']+deletion['minus'+str(l)]
    deletion.drop(['del_'+str(l)+'q','minus'+str(l)], axis=1,inplace=True) 

for l in [17,18]: # frequency of minus > 2%
    deletion['del_'+str(l)+'_merge']=deletion['del_'+str(l)+'p']+deletion['del_'+str(l)+'q']
    deletion.drop(['del_'+str(l)+'p','del_'+str(l)+'q'], axis=1,inplace=True)

for l in [5,7]: # frequency of q and minus > 2% : keep q and sum(p+minus) in a new column and delete the rest    
    deletion['del_'+str(l)+'_merge']=deletion['del_'+str(l)+'p']+deletion['minus'+str(l)]
    deletion.drop(['del_'+str(l)+'p','minus'+str(l)], axis=1,inplace=True)
    
# chromosome 9: freqency of q > 2% : keep q and sum(p+minus) in a new column and delete the rest
deletion['del_9_merge']=deletion['del_9p']+deletion['minus9']
deletion.drop(['del_9p','minus9'], axis=1,inplace=True)

# chromosome x and y: frequency of minusy > 2%: keep column minusy and sum(q+minusx) in new col and delete the rest (p is missing)
deletion['del_x_merge']=deletion['del_xq']+deletion['minusx']
deletion.drop(['del_xq','minusx'],axis=1,inplace=True)

```
